```python
"""
Unzip the folder
"""

from io import BytesIO
import zipfile
unzip = zipfile.ZipFile(BytesIO(streaming_body_1.read()),'r')
file_paths = unzip.namelist()
for path in file_paths:
    unzip.extract(path)
```

```python
"""
Data Augmentation
"""

#Import the ImageDataGenerator library

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
"""
Configure ImageDataGenerator class
"""

#Train_set configuration...

train_datagen = ImageDataGenerator(rescale=1./255,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
```

```python
#Test_set configuration...

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
"""
Applying ImageDataGenerator functionality
"""

#Train_set 
xtrain = train_datagen.flow_from_directory('/home/wsuser/work/dataset/train_set',
                                        target_size=(64,64),
                                        class_mode='categorical',
                                        batch_size=100)
```

```python
#Test_set

xtest = test_datagen.flow_from_directory('/home/wsuser/work/dataset/test_set',
                                        target_size=(64,64),
                                        class_mode='categorical',
                                        batch_size=100)
```

```python
"""
CNN model
"""

#Importing the required library

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D,MaxPooling2D,Flatten,Dense
```

```python
#Initializing the model

model = Sequential()

"""
CNN layers
"""
#Convolutional layer
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(64,64,3)))
#Pooling layer
model.add(MaxPooling2D(pool_size=(2,2)))
#Flatten layer
model.add(Flatten())
#Hidden layers
model.add(Dense(300,activation='relu'))
model.add(Dense(150,activation='relu'))
model.add(Dense(4,activation='softmax'))
```

```python
#Compiling the model

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#Training the model

model.fit(xtrain,
        steps_per_epoch=len(xtrain),
        epochs=20,
        validation_data=xtest,
        validation_steps=len(xtest))
```
